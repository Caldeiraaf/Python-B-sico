# Mostrar a modelagem de um circuito RC (descarga)

```python
import numpy as np
import matplotlib.pyplot as plt
#Descarga de um capacitor C=1, em um resistor R=1
plt.figure(figsize=(10,8))
R=1
C=1
Vm=10
# Constante de tempo
tau=R*C
Fa=100
t= np.arange(0,6*tau,6/Fa)
Vo=Vm*np.exp(-t/tau)
plt.plot(t,Vo,linewidth='4')
plt.title('Descarga do Capacitor')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.text(3,2,'Descarga para R=1 e C=1')
plt.grid(True, which='both')
plt.show()
```
