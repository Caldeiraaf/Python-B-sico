# Cálcule o período fundamental do sinal em tempo discreto

$$x[n]=\cos(\frac{2\pi}{12}n), \\
\omega_{0}~=~\displaystyle \frac{2 \pi}{12}~=~\frac{\pi}{6}, \\
N~=~m \displaystyle \frac{2 \pi}{\omega_{0}}~=~m \displaystyle \Big{(}\frac{2 \pi}{\pi}\Big{)}6~=~m.12 \\
m=1,~N=12$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
n= np.arange(0,16,1)
x=np.cos((np.pi/6)*n)
plt.stem(n,x, use_line_collection=True)
plt.title('Sinal Discreto - Exemplo')
plt.xlabel('Tempo discreto (Amostras)')
plt.ylabel('$x[n]= cos((\pi/6)*n$')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()
```

# Cálcule o período fundamental do sinal em tempo discreto

$$x[n]=\cos(\frac{8\pi}{31}n), \\
\omega_{0}~=~\displaystyle \frac{8 \pi}{31}, \\
N~=~m \displaystyle \frac{2 \pi}{\omega_{0}}~=~m \displaystyle \Big{(}\frac{2 \pi}{8 \pi}\Big{)}31,\\
N~=~m\frac{31}{4} \\
m=4,~N=31$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
n= np.arange(0,40,1)
x=np.cos((8*np.pi/31)*n)
plt.stem(n,x, use_line_collection=True)
plt.title('Sinal Discreto - Exemplo')
plt.xlabel('Tempo discreto (Amostras)')
plt.ylabel('$x[n]= cos((8 \pi/31)*n$')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()
```
