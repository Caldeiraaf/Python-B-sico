## Usando as funções estudadas contrua o sinal:

$$x(t)=3r(t+3)-6r(t+1)+3r(t)-3u(t-3)$$

**Dica:** Defina o vetor de tempo no intervalo entre $-6$ e $6$.

```python
def fun_rampa(t,incl,desl):
    N=len(t)
    r=np.zeros(N)
    for i in range(1,N):
        if t[i] >= -desl:
            r[i] = incl*(t[i]+desl)
    return r
```

```python
def fun_degrau(t,ampl,desl):
    N=len(t)
    u=np.zeros(N)
    for i in range(1,N):
        if t[i] >= -desl:
            u[i] = ampl*1
    return u
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
N=100
t= np.linspace(-6,6,num=N)
xpart=np.zeros([4,N])
xpart[0,:]=fun_rampa(t,3,3)
xpart[1,:]=fun_rampa(t,-6,1)
xpart[2,:]=fun_rampa(t,3,0)
xpart[3,:]=fun_degrau(t,-3,-3)
plt.figure(1)
j=1
while j<=4:
    plt.subplot(4,1,j)
    plt.plot(t,xpart[j-1,:])
    j=j+1
```

```python
x=[sum(i) for i in zip(*xpart)]
#x=xpart[0,:]+xpart[1,:]+xpart[2,:]+xpart[3,:]
plt.figure(2)
plt.figure(figsize=(10,8))
plt.plot(t,x,'r')
plt.xlabel('Tempo (s)')
plt.ylabel('Sinal Completo')
plt.show()
```
