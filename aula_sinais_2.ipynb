# Calcule a energia do sinal de tempo discreto:

$$x[n]~=~  \Bigg{(}\frac{1}{4} \Bigg{)}^{n}u[n]$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
n= np.arange(0,6,0.1)
x=0.25**n
plt.stem(n,x, use_line_collection=True)
plt.title('Sinal Discreto - Exemplo')
plt.xlabel('Tempo discreto (Amostras)')
plt.ylabel('x[n]= 0.25**n')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()
```

# A Energia do Sinal discreto é:

$$ E~=~ \displaystyle \sum\limits_{0 }^{+\infty }\Bigg{|}\Bigg{(}\frac{1}{4} \Bigg{)}^{n}\Bigg{|}^2
    ~=~\sum\limits_{0 }^{+\infty }\Bigg{|}\Bigg{(}\frac{1}{16} \Bigg{)}^{n}\Bigg{|}$$,
                                        
### o resultado é uma série geométrica: portanto, temos:
                                                 
$$E~=~\displaystyle \frac{1}{1-\frac{1}{16}}~=~\frac{16}{15}$$

# Calcule a potência do sinal de tempo discreto:

$$e^{10jn}u[n]$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
n= np.arange(0,6,0.1)
x=np.exp(1j*10*n)
plt.stem(n,x, use_line_collection=True)
plt.title('Sinal Discreto - Exemplo')
plt.xlabel('Tempo discreto (Amostras)')
plt.ylabel('x[n]= e**10jn')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()
```

# A potência do sinal de tempo discreto é:

$$ P~=~ \displaystyle \lim\limits_{N \to \infty } \Bigg{(} \frac{1}{2N+1}  \sum\limits_{n=-N}^{N}|x[n]|^2 \Bigg{)}
    ~=~\displaystyle \lim\limits_{N \to \infty } \Bigg{(} \frac{1}{2N+1}  \sum\limits_{n=0}^{N}1^{2}_{n} \Bigg{)}\\
    ~=~\displaystyle  \lim\limits_{N \to \infty}\frac{N+1}{2N+1}~=~\frac{1}{2}$$
