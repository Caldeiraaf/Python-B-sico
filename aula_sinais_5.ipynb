# A função rampa unitária - r(t)

### A função rampa unitária é zero para valores negativos de $t$ e apresenta uma inclinação unitária para valores positivos de $t$

$r(t) =\left\{\begin{array}{ll}0,  & \quad t < 0 \\ t, & \quad t \geq 0 \end{array}\right.$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-1,2,0.01)
dim=len(t)
r=np.zeros(dim)
for i in range(0,dim) :
    if t[i]>=0:
        r[i]=t[i]

plt.plot(t,r,linewidth='4')
plt.title('Função Rampa unitária')
plt.xlabel('Tempo (s)')
plt.ylabel('r(t)')
plt.grid(True, which='both')
plt.show()
```

## Transformação na função rampa unitária (inclinação e deslocamento)

$$\alpha r(t-\beta)$$

Considerando $\alpha$ a inclinação ($\alpha=1$, rampa unitária) e $\beta$ o deslocamento (atraso, $\beta<0$ e avanço para $\beta>0$). 

```python
def fun_rampa(t,incl,desl):
    N=len(t)
    r=np.zeros(N)
    for i in range(1,N):
        if t[i] >= -desl:
            r[i] = incl*(t[i]+desl)
    return r
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-3,3,0.01)
dim=len(t)
r=np.zeros(dim)
r2=np.zeros(dim)
r=fun_rampa(t,1,1)
r2=fun_rampa(t,1,0)
plt.plot(t,r,'r',t,r2,'--k',linewidth='4')
plt.title('Função Rampa unitária')
plt.xlabel('Tempo (s)')
plt.ylabel('r(t)')
plt.grid(True, which='both')
plt.show()
```

# A função degrau unitário - u(t)

## A função degrau unitário é zero para valores negativos de $t$ e igual $1$ para $t \geq 0$.

$u(t) =\left\{\begin{array}{ll}0,  & \quad t < 0 \\ 1, & \quad t \geq 0 \end{array}\right.$


```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-1,2,0.01)
dim=len(t)
r=np.zeros(dim)
for i in range(0,dim) :
    if t[i]>=0:
        r[i]=1

plt.plot(t,r,linewidth='4')
plt.title('Função Degrau unitário')
plt.xlabel('Tempo (s)')
plt.ylabel('u(t)')
plt.grid(True, which='both')
plt.show()
```

## Transformação na função degrau unitário (amplitude e deslocamento)

$$au(t-b)$$

Considerando $a$ a amplitude ($a=1$, degrau unitário) e $b$ o deslocamento (atraso, $b<0$ e avanço para $b>0$). 

```python
def fun_degrau(t,ampl,desl):
    N=len(t)
    u=np.zeros(N)
    for i in range(1,N):
        if t[i] >= -desl:
            u[i] = ampl*1
    return u
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-3,3,0.01)
dim=len(t)
u=np.zeros(dim)
u2=np.zeros(dim)
u=fun_degrau(t,1,2)
u2=fun_degrau(t,1,0)
plt.plot(t,u,'r',t,u2,'--k',linewidth='4')
plt.title('Função Degrau unitário')
plt.xlabel('Tempo (s)')
plt.ylabel('u(t)')
plt.grid(True, which='both')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
n= np.arange(-5,5,1)
dim=np.size(n) # dim=len(n)
u=np.zeros(dim)
ind=np.where(n>=0)
u[ind]=1
plt.stem(n,u, use_line_collection=True)
plt.title('Função degrau unitário discreta')
plt.xlabel('Tempo (Amostras)')
plt.ylabel('$u(n)$')
plt.grid(True, which='both')
plt.show()
print(dim)
```

# Função impulso unitário - $\delta(t)$

## A função impulso unitário é $1$ para $t=0$ e zero para qualquer valor de $t \neq 0$

$\delta(t) =\left\{\begin{array}{ll}0,  & \quad t \neq 0 \\ 1, & \quad t = 0 \end{array}\right.$

$\displaystyle \int\limits_{-\infty}^{+\infty} \delta(t)~dt~=~1$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
plt.figure(figsize=(8,6))
t= np.arange(-3,3,0.01)
dim=len(t)
#d=signal.unit_impulse(dim,int(dim/2))
d=signal.unit_impulse(dim,'mid')
plt.plot(t,d,linewidth='4')
plt.title('Função impulso unitário')
plt.xlabel('Tempo (s)')
plt.ylabel('$\delta(t)$')
plt.grid(True, which='both')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
n= np.arange(-10,20)
dim=np.size(n) # dim=len(n)
d=np.zeros(dim)
ind=np.where(n==0)
d[ind]=1
plt.stem(n,d, use_line_collection=True)
plt.title('Função impulso unitário discreta')
plt.xlabel('Tempo (Amostras)')
plt.ylabel('$\delta(n)$')
plt.grid(True, which='both')
plt.show()
```
