# O que é o Matplotlib

Para fazer inferências estatísticas necessárias, torna-se necessário visualizar seus dados e o Matplotlib é uma dessas soluções para os usuários do Python. 
É uma biblioteca de plotagem muito poderosa, útil para quem trabalha com Python e NumPy. 
O módulo mais usado do Matplotib é o Pyplot, que fornece uma interface como MATLAB/OCTAVE, mas usa Python e é de código aberto.

# Começando com Pyplot

O Pyplot é um módulo do Matplotlib que fornece funções simples para adicionar elementos de plotagem, como linhas, imagens e texto aos eixos atuais da figura.

## Importando a biblioteca

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Plotando um gráfico simples

````python
fig = plt.figure()
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.show()
````

## Plotando gráficos juntos (na mesma figura)

````python
fig = plt.figure()
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),x,np.cos(x))
plt.show()
````

````python
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2,'--r')
plt.show()
````

````python
# Podemos usar também a sintaxe
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2, color='red', linestyle='dashed')
plt.show()
````

````python
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2, color='red', linestyle='dashed', linewidth='4')
plt.show()
````

````python
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2, color='red', linestyle='dashed', linewidth='4',marker='o',
         markerfacecolor='blue', markersize=6)
plt.show()
````

````python
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

## Títulos e Grid

````python
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2, color='red', linestyle='dashed', linewidth='4',marker='o',
         markerfacecolor='blue', markersize=6)
plt.title('$y=t^2$') # Vocẽ pode usar código Latex
plt.grid()
plt.show()
````

## Rótulos dos eixos (labels)

````python
fig = plt.figure()
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2, color='red', linestyle='dashed', linewidth='4',marker='o',
         markerfacecolor='blue', markersize=6)
plt.title('$y=t^2$') # Vocẽ pode usar código Latex
plt.xlabel('Tempo (s)')
plt.ylabel('y')
plt.grid()
plt.show()
````

## Controlando o tamanho da figura

````python
fig = plt.figure(figsize=(12,8))
t=np.arange(0,5,0.2) # tempo de amostragem uniforme de 200ms
plt.plot(t,t**2, color='red', linestyle='dashed', linewidth='4',marker='o',
         markerfacecolor='blue', markersize=6)
plt.title('$y=t^2$') # Vocẽ pode usar código Latex
plt.xlabel('Tempo (s)')
plt.ylabel('y')
plt.grid()
plt.show()
````

## Múltiplos gráficos 

````python
plt.figure()
x=np.linspace(0,10,100)
plt.subplot(2, 1, 1) # (linhas, colunas, número do painel)
plt.plot(x, np.sin(x),'k')
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x),'r');
````

````python
fig,ax = plt.subplots(2)
x=np.linspace(0,10,100)
ax[0].plot(x, np.sin(x),'k')
ax[1].plot(x, np.cos(x),'r');
````
