<h1 align=center> Sinais Contínuos <h1>


```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(0,3,0.01)
f=1
y=np.sin(2*np.pi*f*t)
plt.plot(t,y)
plt.title('Sinal Senoidal')
plt.xlabel('Tempo (s)')
plt.ylabel('y(t)= sin(t)')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()

 
```

<h1 align=center> Sinais Discretos <h1>

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
n= np.arange(0,3,0.1)
f=1
y=np.sin(2*np.pi*f*n)
plt.stem(n,y, use_line_collection=True)
plt.title('Sinal Senoidal')
plt.xlabel('Tempo discreto (Amostras)')
plt.ylabel('y[n]= sin[n]')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()
```

# Sinais de energia - Sinais Contínuos

Para um sinal x(t), a Energia é definida como:
    
$$ E~=~ \displaystyle \int\limits_{t_{1}}^{t_{2}}|x(t)|^2 ~dt $$

A Energia total

$$ E_{\infty}~=~ \displaystyle \int\limits_{-\infty }^{+\infty }|x(t)|^2~dt $$

# Sinais de Energia - Sinais Discretos

Para um sinal x[n], a Energia é definida como:
    
$$ E~=~ \displaystyle \sum\limits_{n1}^{n2}|x[n]|^2 $$

A Energia total

$$ E_{\infty}~=~ \displaystyle \sum\limits_{-\infty }^{+\infty }|x[n]|^2 $$

# Sinais de Potência - Sinais Contínuos

Para um sinal x(t), a Potência é definida como:
    
$$ P~=~ \displaystyle \frac{1}{t_{2}-t_{1}} \int\limits_{t_{1}}^{t_{2}}|x(t)|^2 ~dt $$

A Potência total

$$ P_{\infty}~=~ \displaystyle \lim\limits_{T \to \infty } \frac{1}{2T} \int\limits_{-T}^{T}|x(t)|^2~dt $$


# Sinais de Potência - Sinais Discretos

Para um sinal x[n], a Energia é definida como:
    
$$ P~=~ \displaystyle \frac{1}{n_{2}-n_{1}} \sum\limits_{n_{1}}^{n_{2}}|x[n]|^2 $$

A Potência total

$$ P_{\infty}~=~ \displaystyle \lim\limits_{N \to \infty } \frac{1}{2N+1}  \sum\limits_{n=-N}^{N}|x[n]|^2 $$

<h1 align=center> Exemplo do cálculo da energia do sinal contínuo <h1>

# Dado um sinal definido como:

$$x(t) = e^{-t}u(t)$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(0, 7, 0.01)
f=1
y=np.exp(-t)
plt.plot(t,y)
plt.title('Sinal Exponencial')
plt.xlabel('Tempo (s)')
plt.ylabel('y(t)= e(-t)')
plt.grid(True, which='both')
plt.axhline(y=0, color='r')
plt.show()
```

# A energia do sinal é:

$$ E~=~ \displaystyle \int\limits_{-\infty }^{+\infty }|e^{-t}u(t)|^2~dt
~=~ \int\limits_{0}^{+\infty }|e^{-t}|^2~dt
~=~-\frac{1}{2}e^{-2t}\Big{|}_{0}^{\infty}~=\frac{1}{2}$$
