# Operações básicas (revisão)
- Para um sinal $x(t)$, verificar:
    - $a~x(t)$
    - $x(t)+a$
    
Considerar valores positivos e negativos para a.


```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-1,5,0.01)
dim=len(t)
x=np.zeros(len(t))
for i in range(-1,dim) :
    if t[i]>0 and t[i]<=1:
        x[i]=1
    elif t[i]>1 and t[i]<=2:
        x[i]=-t[i]+2
    else:
        x[i]=0
a=-0.5
plt.plot(t,x,'r',t,x+a,'--k',linewidth='5')
plt.title('Sinal Exemplo - Transformações básicas')
plt.xlabel('Tempo (s)')
plt.ylabel('x(t)')
plt.grid(True, which='both')
plt.show()
```

# Transformação no argumento de $x(t) ~\to~ x(\alpha t + \beta)$

- ## Deslocamento no tempo: $\alpha=1$, $\beta \neq 0$ 
    
    - ### Atraso ($\beta < 0$)
    
        - ### Para $\beta=-0,5$, temos:
    
        $$x(t-0,5)$$
        
    - ### Avanço ($\beta > 0$)
        
        - ### para $\beta = 1$, temos:
        
        $$x(t+1)$$
    
    


```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-1,5,0.01)
dim=len(t)
x=np.zeros(len(t))
for i in range(-1,dim):
    if t[i]>0 and t[i]<=1:
        x[i]=1
    elif t[i]>1 and t[i]<=2:
        x[i]=-t[i]+2
    else:
        x[i]=0
beta=1
m=t-beta

        
plt.plot(t,x,'r',m,x,'--k',linewidth='5')
plt.title('Sinal Exemplo - Transformações no argumento (atraso e avanço) ')
plt.xlabel('Tempo (s)')
plt.ylabel('x(t)')
plt.grid(True, which='both')
plt.show()
```

## Reflexão no tempo (Inversão): $\alpha=-1$, $\beta = 0$ 


- ###  $ \alpha = -1, \to x(t) = x(-t) $ - Inversão do sinal
    
    

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-1,5,0.01)
dim=len(t)
x=np.zeros(len(t))
for i in range(0,dim):
    if t[i]>0 and t[i]<=1:
        x[i]=1
    elif t[i]>1 and t[i]<=2:
        x[i]=-t[i]+2
    else:
        x[i]=0
plt.plot(t,x,'r',-t,x,'--k',linewidth='5')
plt.title('Sinal Exemplo - Transformações no argumento (inversão) ')
plt.xlabel('Tempo (s)')
plt.ylabel('x(t)')
plt.grid(True, which='both')
plt.show()
```

## Mudança de escala de tempo: $0<\alpha \neq 1$, $\beta = 0$ 

- ### $\alpha=2 \to x(2t)$ - Compressão do sinal

- ### $\alpha=\frac{1}{2} \to x(\frac{t}{2})$ - Expansão do sinal 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
t= np.arange(-2,2,0.01)
dim=len(t)
x=np.zeros(len(t))
for i in range(-2,dim):
    if t[i]>-1 and t[i]<=0:
        x[i]=t[i]+1
    elif t[i]>0 and t[i]<=1:
        x[i]=-t[i]+1
    else:
        x[i]=0
# No meu programa ficou invertido
plt.plot(t,x,'r',t/2,x,'--k',linewidth='5') # compressão
#plt.plot(t,x,'r',2*t,x,'--k',linewidth='5') # expansão
plt.title('Sinal Exemplo - Transformações no argumento (escalamento) ')
plt.xlabel('Tempo (s)')
plt.ylabel('x(t)')
plt.grid(True, which='both')
plt.show()
```
