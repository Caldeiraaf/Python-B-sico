# Sistemas com e sem memória (instantâneo e dinâmico)

### Um sistema é dito sem memória ou instantâneo se a saída em qualquer instante de tempo depender apenas da entrada naquele instante de tempo. Caso contrário, o sistema é com memória ou dinâmico.
$$y[n]~=~x[n]+1 \quad \to \mbox{Sistema sem memória}$$


$$y[n]~=~2x[n]+x[n-3] \quad \to \mbox{Sistema com memória}$$
$$
y(t)~=~\displaystyle \frac{1}{C}\int\limits_{-\infty}^{t} x(\tau) \, d\tau \quad \to \mbox{Sistema com memória}
$$

# Sistemas causal e não causal

### Um sistema é causal (também conhecido como físico ou não antecipativo) se o valor atual da saída depender somente dos valores presentes e/ou passados do sinal de entrada. Se depender  de valores futuros, o sistema é não causal.

$$ y[n]~=~x[n]+x[n+1] \quad \to \mbox{Sistema não causal}$$

$$y(t)~=~x(t)+\cos(t+2) \quad \to \mbox{Sistema causal}$$

$$y[n]~=x[-n] \quad \to \mbox{Sistema não causal}$$

# Estabilidade

### Estabilidade é uma propriedade importante dos sistemas. Informalmente, um sistema estável é aquele em que pequenas entradas levam a respostas que não são divergentes.

# Estabilidade BIBO (bounded input / bounded output)

### Um sistema é BIBO estável se toda entrada limitada resultar em uma saída limitada.

$$y[n]~=~nx[n+1], ~ \mbox{para}~n\geq 0 \quad \to \mbox{Sistema instável}$$

$$y(t)~=~x(t)\cos(t) \quad \to \mbox{Sistema estável}$$

$$y[n]~=~10x[n] \quad \to \mbox{Sistema estável}$$

# Sistemas lineares

## O sistema é dito linear quando é válido o princípio da superposição:

- Aditividade
    - A resposta do sistema a $x_{1}(t)~+~x_{2}(t)$ é $y_{1}(t)~+~y_{2}(t)$
- Homogeneidade
    - A resposta do sistema a $ax_{1}(t)$ é $ay_{1}(t)$, em que $a$ é qualquer constante complexa. 

$$y[n]~=~2x[n]-1 \quad \to \mbox{Não Linear} \\
2ax[n]-1~~\neq~~a(2x[n]-1) \quad (\mbox{falhou no teste da homogeneidade})$$

$$y(t)~=~tx(t) \quad \to \mbox{Linear} \\
tax(t)~~=~~a(tx(t)) \quad (\mbox{homogeneidade ok}) \\
t(x_{1}(t)+x_{2}(t))~~=~~tx_{1}(t)+tx_{2}(t) \quad (\mbox{aditividade ok})
$$
